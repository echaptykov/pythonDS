Задание 3

```python
x = np.linspace(-3,3,51,True)
```

```python
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)

```

```python
fig, Axes = plt.subplots(2,2)
fig.set_size_inches(8,6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

Axes[0,0].plot(x, y1)
Axes[0,0].set_title("График y1")
Axes[0,0].set_xlim([-5, 5])

Axes[0,1].plot(x, y2)
Axes[0,1].set_title("График y2")

Axes[1,0].plot(x, y3)
Axes[1,0].set_title("График y3")

Axes[1,1].plot(x, y4)
Axes[1,1].set_title("График y4")
```

```python
V1_class1 = df[['V1','Class']]
```

```python
V1_class1 = V1_class1.set_index('Class')
```

```python
V1_class0 = V1_class1.filter(like = '0',axis = 0)
```

```python
V1_class1 = V1_class1.filter(like = '1',axis = 0)
```
